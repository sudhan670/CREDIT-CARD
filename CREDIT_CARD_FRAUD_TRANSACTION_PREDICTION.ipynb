```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card=pd.read_csv('/content/creditcard.csv')
```

```python
# separating the data for analysis
legit=credit_card[credit_card.Class == 0] # class value is 0 all values are stored in legit
fraud=credit_card[credit_card.Class == 1] # Class value is 1 all values are stroed in fraud
```

this datasets are highly unbalanced so 0 represents normal transactions and 1 representd fraudlecent transacions

```python
print(legit.shape)
print(fraud.shape)
```

```python
fraud.Amount.describe()
```

```python
# compare values for both transactions
credit_card.groupby('Class').mean()
```

Under-Sampling to build sample datasets containing similar datasets of legit transactions

the no of fraudlenct transactions is 52

```python
legit_sample = legit.sample(n=52)
```

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0) # axis 0 because data add one by one
```

```python
new_dataset.head()
```

```python
new_dataset.tail() # fraudlenct transactions
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting data features and targets 

```python
X = new_dataset.drop(columns='Class', axis=1) # axis 1 repesent columns
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into training and split data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Modelling training Logistic Regression for Binary regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model with Training data
model.fit(X_train, Y_train) # Y_train contains corressponding label for those data
```

Model Evalution Accuracy Score train data

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data:', training_data_accuracy )
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('Accuracy score on Test Data: ', test_data_accuracy)
```
